```python
def dictdiff(first, second):
    output = {}
    for one_key in first.keys() | second.keys():
        v1 = first.get(one_key)
        v2 = second.get(one_key)
        
        if v1 != v2:
            output[one_key] = [v1, v2]
        
    return output

d1 = {'a': 1, 'b': 2, 'c': 3}
d2 = {'a': 1, 'b': 2, 'c': 4}

print(dictdiff(d1, d1))
# prints {}

print(dictdiff(d1, d2))
# # # # # # prints {'c': [3, 4]}

d1 = {'a': 1, 'b': 2, 'd': 3}
d2 = {'a': 1, 'b': 2, 'c': 4}

print(dictdiff(d1, d2))
# # # # # # prints {'c': [None, 4], 'd': [3, None]}

d1 = {'a': 1, 'b': 2, 'c': 3}
d2 = {'a': 1, 'b': 2, 'd': 4}

print(dictdiff(d1, d2))
# # # # prints {'c': [3, None], 'd': [None, 4]}
```

```python
# write a function "firstlast" that takes a string, list, or tuple
# of any length.

# it'll return the same data type with only two elements —
# taken from the first and last elements of the input.

# SO if I call:

def firstlast(data):
#    return data[:1] + data[-1:]
     return data[::len(data)-1]
    

print(firstlast('abcd'))  # 'ad'
print(firstlast([10, 20, 30, 40]))   # [10, 40]
print(firstlast((30, 50, 60, 70, 80)))  # (30, 80)
```

```python
# write a function "all_lines"
# - mandatory argument for "filename", the name of the file
#   into which we'll write the output
# - any number of filenames after that, which will be our
#   inputs
# 
# write each line from the input files into the output file,
# prefacing each line with the file count and the line count
# (line count is specific to that file)

# [0][0] first line of first file
# [0][1] second line of first file
# [1][0] first line of second file

def all_lines(outfilename, *args):
    with open(outfilename, 'w') as outfile:
        for one_filename in args:
            for one_line in open(one_filename):
                outfile.write(one_line)

all_lines('output.txt', 
          '/etc/passwd', 
         '/Users/reuven/Consulting/LinuxJournal/articles/2018-12-pathlib.txt',
         '/Users/reuven/Consulting/LinuxJournal/articles/2019-04-mypy.txt')

```
